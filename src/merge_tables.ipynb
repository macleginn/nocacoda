```python
import os.path
from glob import glob
import pandas as pd
```

```python
data_dir = 'data'
lang_files = sorted(glob(os.path.join('..', data_dir, '*_data.csv')))
marker_files = sorted(glob(os.path.join('..', data_dir, '*_markers.csv')))
assert len(lang_files) == len(marker_files)
```

```python
list(zip(lang_files, marker_files))
```

```python
lang_data = pd.read_csv(lang_files[0], sep='\t')
lang_data.head()
```

```python
marker_data = pd.read_csv(marker_files[0], sep='\t')
marker_data.head()
```

```python
sentences = pd.read_csv(os.path.join('..', data_dir, '03_sentences.csv'), sep='\t')
languages = pd.read_csv(os.path.join('..', data_dir, '01_languages.csv'), sep='\t')
```

```python
languages.head()
```

```python
sentences.head()
```

```python
lang_dfs = []
for lang_file, marker_file in zip(lang_files, marker_files):
    print(lang_file, marker_file)
    assert lang_file.rsplit(os.path.sep, 1)[-1].split('_')[0] == marker_file.rsplit(os.path.sep, 1)[-1].split('_')[0]
    lang_data = pd.read_csv(lang_file, sep='\t')
    marker_data = pd.read_csv(marker_file, sep='\t')
    if 'explicitness' not in marker_data.columns:
        marker_data['explicitness'] = pd.NA
    sent_w_marker_data = pd.merge(left = lang_data, right = marker_data[[
        'causal_marker',
        'morphosyntactic_type',
        'explicitness',
        'goal',
        'instrument',
        'location',
        'object',
        'source', 
        'other_details',
        'other'
    ]], how = 'left', on = 'causal_marker')
    sent_w_all_data = pd.merge(left = sent_w_marker_data, right = sentences[[
        'sentence_no',
        'type',
        'objective_subjective',
        'direct_indirect',
        'internal_external',
        'taxis'
    ]], how = 'left', left_on = 'stimulus_no', right_on = 'sentence_no')
    del sent_w_all_data['sentence_no']
    lang_dfs.append(sent_w_all_data)
```

```python
all_data = pd.concat(lang_dfs)
all_data.shape
```

```python
all_data.to_csv('combined_data.csv', index=False, sep='\t')
```
